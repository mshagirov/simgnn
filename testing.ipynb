```python
from graphviz import Digraph
```

```python
dot = Digraph(comment='The Round Table',graph_attr={'rankdir':'LR'},
             node_attr={'color': 'lightblue2'})

dot.node('A', label='''<<TABLE><TR>
<TD>&sum;w<sub>i</sub>x<sub>i</sub></TD>
<TD>f<sub>a</sub>()</TD>
</TR> </TABLE>>''')
dot.node('B',label='''<x<sub>0</sub>>''',shape='circle',style= 'filled')
dot.node('L', '''<&alpha;>''',style= 'filled')
dot.node('D','''<&#8942;>''')
dot.edges(['AB', 'AL', 'AD'])
dot.edge('B', 'L', constraint='false')
# dot.render('round-table.gv',format='png')
dot
```

```python
import graphviz

d = graphviz.Digraph(filename='NN.gv',graph_attr={'rankdir':'LR'},
                     node_attr={'color': 'lightblue2','shape':'circle','style':'filled'},format='png')

with d.subgraph(name='l0') as l0:
    l0.attr(rank='same',style='dotted')
    l0.node('x0','''<x<sub>0</sub>>''')
    l0.node('x1','''<x<sub>1</sub>>''')
    l0.node('x2','''<x<sub>2</sub>>''')
    l0.node('xs','''<&#8942;>''')
    l0.node('xn','''<x<sub>n</sub>>''')
    l0.edges([('x0','x1'),('x1','x2'),('x2','xs'),('xs','xn')])
    l0.edge_attr['style'] = 'invis'
    l0.attr(label='input')

with d.subgraph(name='l1') as l1:
    l1.attr(rank='same', style='dotted')
    l1.node('a0','''<a<sub>0</sub>>''')
    l1.node('a1','''<a<sub>1</sub>>''')
    l1.node('a2','''<a<sub>2</sub>>''')
    l1.node('a3','''<a<sub>3</sub>>''')
    l1.edges([('a0','a1'),('a1','a2'),('a2','a3')])
    l1.edge_attr['style'] = 'invis'
    l1.node_attr['color'] = 'lightgrey'
    l1.attr(label='hidden 1')
with d.subgraph(name='w1') as w1:
    w1.attr(overlap='true',concentrate='true')
    for x_l in ['x0','x1','x2','xs','xn']:
        for h_l in ['a0','a1','a2','a3']:
            w1.edge(x_l,h_l)
#     w1.edge_attr['color']='blue'
    w1.edge_attr['style']='solid'
```
